<font color='red'>Définition inversibilité<br>

Soit $e$ un élément neutre, on dit que $a$ est inversible à gauche s'il admet un élément $a'$ tel que $a'*a=e$. On dit que $a$ est inversible à droite s'il admet un élément $a'$ tel que $a*a'=e$. $a$ est inversible s'il est inversible à gauche et à droite.<br>

Dans le cas des matrices, on dit que $M$ est inversible s'il existe une matrice $M^{-1}$ tel que $MM^{-1}=M^{-1}M=I$<br></font>

Implications:<br>
$M \in M_n(K)$, $M$ inversible $<=>$ $M$ est de plein rang $<=>$ déterminant non nul $<=>$ $Ker(M)=\{0\}$<br>
$M$ inversible $<=>$ $M$ definie positive, ie toutes ses valeurs propres sont strictement positives (démo Algèbre 2)

Algorithmes de Gauss: toute matrice inversible peut être transformée en la matrice identité avec des opérations élémentaires sur les lignes ou les colonnes.

Une matrice est singulière si au moins une colonne (ou ligne) est une combinaison linéaire de ses autres colonnes (ou lignes).<br>
Matrice régulière = matrice inversible = matrice non singulière

Un scalaire $\lambda$ est une valeur propre de $f$ s'il existe un vecteur $u$ non nul tel que $f(u) = λu$.

$M$ semi-définie positive $<=>$ $\lambda_i \ge 0~\forall{i}$<br>

Soit $A$ une matrice semi-définie positive et $X$ un vecteur propre non nul:<br>
