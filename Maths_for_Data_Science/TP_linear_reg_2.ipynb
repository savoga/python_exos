# 4) ACP

Problème, en gardant un grand nombre de variables explicatives, $(X^TX)^{-1}$ va être très dur à calculer<br>
On veut déterminer quelles sont les variables les plus importantes à garder --> ACP

<u>Rappel de la méthode ACP</u>:

Expression de la matrice de variance-covariance centrée $\Sigma_n$

Décomposition spectrale de $\Sigma_n$

Analyse des valeurs propres (spectre)

On ne conserve que la dimension pour laquelle les valeurs propres sont les plus grandes --> ce sont celles qui expliquent le mieux la variance

On réécrit la matrice de variance-covariance avec la dimension finale

"Comme les variables sont exprimées avec des unités différentes, on effectue l'ACP sur données centrées réduites, ce qui conduit à chercher les valeurs et vecteurs propres de la matrice de corrélation"

# 5) Régression linéaire

On recréé une matrice de variables explicatives avec un spectre des plus grandes valeurs propres

```python
eig_vec_60 = eig_vec[-60:]
```

```python
Y_train, Y_test = train_test_split(Y, test_size=0.2)
```

ATTENTION: ici les données ne sont pas centrées alors que normalement on fait une ACP sur des données centrées (cf cours), cependant on utilisera l'intercept dans la régression

```python
X_train.shape
```

```python
eig_vec_60.shape
```

```python
X_tilde = np.dot(X_train, eig_vec_60.T) # La nouvelle matrice X à considérer
```

```python
X_tilde.shape
```

```python
from sklearn.linear_model import LinearRegression
reg_pca = LinearRegression(fit_intercept=True).fit(X_tilde, Y_train)
```

```python
xxx = np.arange(0, len(reg_pca.coef_[0]))
```

```python
plt.plot(xxx, (np.sort(reg_pca.coef_)).T)
plt.title("Coefficients de la régression après ACP")
```

```python
reg = LinearRegression(fit_intercept=True).fit(X_train, Y_train)
```

```python
xxxx = np.arange(0, len(reg.coef_[0]))
```

```python
plt.plot(xxxx, (np.sort(reg.coef_)).T)
plt.title("Coefficients de la régression sans ACP")
```
