# 4) ACP

Problème: en gardant un grand nombre de variables explicatives, $(X^TX)^{-1}$ va être très dur à calculer<br>
On veut déterminer quelles sont les variables les plus importantes à garder --> ACP

<u>Rappel de la méthode ACP</u>:

Expression de la matrice de variance-covariance centrée $\Sigma_n$

Décomposition spectrale de $\Sigma_n$

Analyse des valeurs propres (spectre)

On ne conserve que la dimension pour laquelle les valeurs propres sont les plus grandes --> ce sont celles qui expliquent le mieux la variance

On réécrit la matrice de variance-covariance avec la dimension finale

"Comme les variables sont exprimées avec des unités différentes, on effectue l'ACP sur données centrées réduites, ce qui conduit à chercher les valeurs et vecteurs propres de la matrice de corrélation"

# 8) Forward selection

```python
from scipy.stats import t
from scipy.stats import norm
import operator
```

```python
def forwardSelection(X, Y, nb_var, p_value_threshold):
    Y_temp=np.array(Y).reshape(-1,1)
    p_values_sel=[]
    T_stats_sel=[]
    idx_rem_var=list(range(nb_var))
    select_var_idx=[]
    res=[]
    T_stats_trace = np.zeros((p,p))
    pval_trace = np.zeros((p,p))

    p_values={}
    T_stats={}
    res={}
    # On boucle sur toutes les variables
    for i in range(nb_var):
    
        p_values.clear()
        T_stats.clear()
        res.clear()
        # On boucle sur toutes les variables restantes après sélection
        for j in idx_rem_var:
            X_j=np.array(X.iloc[:,j]).reshape(-1,1)
            n=len(X_j)
            X_j_intercept=np.hstack((np.ones((n,1)),X_j))
        
            # Régression
            reg=LinearRegression(fit_intercept=True).fit(X_j, Y_temp)
            res_j=reg.predict(X_j)-Y_temp
            res[j]=res_j
        
            # T-stat
            sigma_est_squarred=np.sum(res_j ** 2)/(n-2) # Rang = 2
            sigma_est=np.sqrt(sigma_est_squarred)
            XtX = np.dot(X_j_intercept.T,X_j_intercept)
            XtX_inv = np.linalg.inv(XtX)
            T_j=np.abs(reg.coef_[0])/(sigma_est*np.sqrt(XtX_inv[1][1]))
            T_stats[j]=T_j
            T_stats_trace[i,j]=T_j
        
            # p-value
            p_value=2*(1-t.cdf(T_j,n-2))
            p_values[j]=p_value
            pval_trace[i,j]=p_value
    
        idx_max = max(T_stats.items(), key=operator.itemgetter(1))[0]
        Y_temp=res[idx_max]
        
        idx_rem_var=np.setdiff1d(idx_rem_var,idx_max)
        if(p_values[idx_max]>p_value_threshold):
            break
        select_var_idx.append(idx_max)
        T_stats_sel.append(T_stats[idx_max])
        p_values_sel.append(p_values[idx_max])
    return select_var_idx, T_stats_sel, p_values_sel, T_stats_trace, pval_trace
```

```python
select_var_idx, T_stats_sel, p_values_sel, T_stats_trace, pval_trace=forwardSelection(X_train, Y_train, p, 1)
```
