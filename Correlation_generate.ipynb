```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
num_samples = 400

# we want standardized variables in order to use correlation matrix
# and not cov matrix

mu = np.array([0.0, 0.0])
variances = 1

corr = 0.3

corr_mat = np.array([
        [1, corr],
        [corr, 1]
        ])

y = np.random.multivariate_normal(mu, corr_mat, size=num_samples)

plt.plot(y[:,0], y[:,1], 'b.')
plt.show()

print(np.corrcoef(y[:,0], y[:,1]))
```
