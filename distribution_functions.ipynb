```python
import numpy as np

import matplotlib.pyplot as plt
```

## Normal

```python
data = np.random.normal(0,1,10000)
plt.hist(data)
plt.title('Normal distribution (mu=0, sigma=1, n=10000)')
plt.show()
```

## Binomial

```python
data = np.random.binomial(100,0.2, size=10000)
plt.hist(data)
plt.title('Binomial distribution (n_trials=100, proba=0.2, size=10000)')
plt.show()
```

## Beta

```python
data = np.random.beta(4,6,size=10000)
plt.hist(data)
plt.title('Beta distribution (a=4, b=6, size=10000)')
plt.show()
```

## Pareto

```python
data = np.random.pareto(10, size=100)
plt.hist(data)
plt.title('Pareto distribution (a=2, size=100)')
plt.show()
```
