```python
# 1.
import pandas as pd
data = pd.read_csv("https://bitbucket.org/portierf/shared_files/downloads/Galton.txt", sep='\t')
```

```python
data
```

```python
# 2.
data["MeanParents"] = 0.5*(data["Father"] + 1.08*data["Mother"])
```

```python
data
```

```python
# 3.
import matplotlib.pyplot as plt

x = data["MeanParents"]
y = data["Height"]

fig = plt.figure(figsize=(12,12))
plt.scatter(x,y)
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants') 
plt.grid(True)
plt.show()
```

On voit que la taille des enfants est linÃ©aire avec la taille des parents

```python
#4.
import numpy as np
from sklearn.linear_model import LinearRegression
x = np.array(x).reshape(-1,1)
# Note: sans reshape, np.array(x) renvoie un vecteur de nombres --> .fit() requiert un vecteur d'array
y = np.array(y).reshape(-1,1)

# fit:
# trouve les coefficients optimaux (la mÃ©thode d'optimisation utilisÃ©e n'est pas l'inversion)
# = Ã©tape d'entraÃ®nement
# on utilise 'fit_intercept' car les donnÃ©es ne sont pas centrÃ©es et on veut la constante
reg = LinearRegression(fit_intercept=True).fit(x, y)

# predict:
# utilise les coefficients estimÃ©s pour calculer Y
y_pred_1 = reg.predict(x)

fig = plt.figure(figsize=(12,12))
plt.scatter(x,y)
plt.scatter(x,y_pred_1)
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()
```

<h1><center><i>fit()</i> (cas multivariÃ©)</center></h1>
$$(X^TX)\widehat{\theta}=X^TY$$

<h1><center><i>predict()</i> (cas multivariÃ©)</center></h1>
$$pred(X)=X\widehat{\theta}$$

La rÃ©gression calculÃ©e donne une droite qui passe par le milieu du nuage de points.

```python
# 5.
# Recentrage: on retranche la moyenne de toutes les observations pour la taille des parents et celle des enfants
y_c = y - y.mean()
x_c = x - x.mean()

# On calcule les coefficients de la rÃ©gression dans le cas centrÃ©
reg_c = LinearRegression(fit_intercept=False).fit(x_c, y_c)

y_pred_c = reg_c.predict(x_c)

# On compare les rÃ©sultats obtenus avec les rÃ©sultats prÃ©cÃ©dents Ã  l'aide de la formule (cf plus bas)

# Comparaison des coefficients
# Cas non centrÃ©
theta_0 = reg.intercept_
theta_1 = reg.coef_
# Cas centrÃ©
theta_0_c = y.mean() - x.mean()*reg_c.coef_
theta_1_c = reg_c.coef_

print("Dans le cas non centrÃ©, theta0 = %f" % theta_0[0])
print("Dans le cas centrÃ©, theta0 = %f" % theta_0_c[0][0])
print("")
print("Dans le cas non centrÃ©, theta1 = %f" % theta_1[0][0])
print("Dans le cas centrÃ©, theta1 = %f" % theta_1_c[0][0])

# Comparaison des prÃ©dictions
y_pred_centre=reg_c.predict(x_c)
np.allclose(y_pred_centre+y.mean(),y_pred_1)
```

On voit qu'on rÃ©cupÃ¨re bien les mÃªmes rÃ©sultats d'optimisation dans le cas centrÃ© et non centrÃ© grÃ¢ce Ã  la formule vue en cours:

$$\widehat{\theta}=
\quad
\begin{pmatrix}
\overline{Y}-\overline{\tilde{X^T}}\tilde{\theta} \\
\tilde{\theta}
\end{pmatrix}
\quad
$$

```python
# 6.
res = y - y_pred_1
plt.hist(res)
plt.title('RÃ©sidus', fontsize=10)
plt.show()
```

$$r_{i}=y_{i}-\widehat{y_{i}}$$

L'hypothÃ¨se de normalitÃ© ne semble pas trÃ¨s crÃ©dible au vu de la forme de la courbe (la courbe n'est pas trÃ¨s en cloche).

On peut aussi calculer le kurtosis pour mieux dÃ©crire la forme de la courbe.
Ce kurtosis Ã©tant trÃ¨s faible (nÃ©gatif), cela signifie que la distribution est bien plus aplatie que celle de la loi normale:

```python
from scipy.stats import kurtosis
print(kurtosis(res))
```

```python
# 7.
x_filles = np.array(data[data["Gender"] == "F"]["MeanParents"]).reshape(-1,1)
y_filles = np.array(data[data["Gender"] == "F"]["Height"]).reshape(-1,1)

x_garcons = np.array(data[data["Gender"] == "M"]["MeanParents"]).reshape(-1,1)
y_garcons = np.array(data[data["Gender"] == "M"]["Height"]).reshape(-1,1)


fig = plt.figure(figsize=(12,12))
plt.scatter(x,y_pred_1)
plt.scatter(x_filles,y_filles,c='coral')
plt.scatter(x_garcons,y_garcons,c='lightblue')
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()
```

On voit qu'il y a de fortes disparitÃ©s entre la taille des garÃ§ons et la taille des filles.
Ces diffÃ©rences crÃ©ent des Ã©carts importants entre les valeurs rÃ©elles et celles prÃ©dites, d'oÃ¹ la forme Ã©paisse de la courbe des rÃ©sidus.

```python
# 8.
reg_garcons = LinearRegression(fit_intercept=True).fit(x_garcons, y_garcons)
y_pred_garcons_1 = reg_garcons.predict(x_garcons)

fig1 = plt.figure(figsize=(12,12))
plt.scatter(x,y_pred_1)
plt.scatter(x_filles,y_filles,c='coral')
plt.scatter(x_garcons,y_garcons,c='lightblue')
plt.scatter(x_garcons,y_pred_garcons_1,c='orange')
plt.xlabel('Taille parents')
plt.ylabel('Taille enfants')
plt.show()

res_garcons = y_garcons - y_pred_garcons_1
fig2 = plt.figure(figsize=(12,12))
plt.hist(res_garcons)
plt.title('RÃ©sidus rÃ©gression garÃ§ons', fontsize=10)
plt.show()
```

On remarque que la distribution des rÃ©sidus ressemble davantage Ã  une courbe gaussienne lorsque l'on prend des individus garÃ§ons uniquement. On peut vÃ©rifier cette information en calculant le kurtosis, plus Ã©levÃ© quÃ© prÃ©cÃ©demment:

```python
print(kurtosis(res_garcons))
```

```python
# 9.
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(data, test_size=0.2)

# Echantillon de test
x_test_garcons = np.array(data_test[data_test["Gender"]=="M"]["MeanParents"]).reshape(-1,1)
y_test_garcons = np.array(data_test[data_test["Gender"]=="M"]["Height"]).reshape(-1,1)

# Echantillon d'apprentissage
# ModÃ¨le 1 - garÃ§ons uniquement
x_train_garcons = np.array(data_train[data_train["Gender"] == "M"]["MeanParents"]).reshape(-1,1)
y_train_garcons = np.array(data_train[data_train["Gender"] == "M"]["Height"]).reshape(-1,1)
reg_train_garcons = LinearRegression(fit_intercept=True).fit(x_train_garcons, y_train_garcons)
y_pred_garcons_2 = reg_train_garcons.predict(x_test_garcons)
# ModÃ¨le 2 - Ã©chantillon complet
x_train_all = np.array(data_train["MeanParents"]).reshape(-1,1)
y_train_all = np.array(data_train["Height"]).reshape(-1,1)
reg_train_all = LinearRegression(fit_intercept=True).fit(x_train_all, y_train_all)
y_pred_2 = reg_train_all.predict(x_test_garcons)

N = len(y_test_garcons)

rPred_garcons = 0
for i in range(N):
    rPred_garcons+=((y_test_garcons[i] - y_pred_garcons_2[i])**2)
rPred_garcons = rPred_garcons/N

rPred_all = 0
for i in range(len(x_test_garcons)):
    rPred_all+=((y_test_garcons[i] - y_pred_2[i])**2)
rPred_all = rPred_all/N

print("Le risque de prÃ©diction du modÃ¨le basÃ© sur les garÃ§ons est %f" % rPred_garcons)
print("Le risque de prÃ©diction du modÃ¨le basÃ© sur l'Ã©chantillon complet est %f" % rPred_all)
```

On constate que le risque prÃ©dictif est plus faible dans le cas d'une prÃ©diction sur les garÃ§ons seulement. Ce rÃ©sultat est attendu car, comme vu Ã  la question 8, les erreurs rÃ©siduels sont plus faibles lorsque le calcul des coefficients est fait sur l'Ã©chantillon des garÃ§ons uniquement.

```python
# 10.
x_test_filles = np.array(data_test[data_test["Gender"]=="F"]["MeanParents"]).reshape(-1,1)
y_test_filles = np.array(data_test[data_test["Gender"]=="F"]["Height"]).reshape(-1,1)

# On rajoute un modÃ¨le: ModÃ¨le 3 - filles uniquement
x_train_filles = np.array(data_train[data_train["Gender"] == "F"]["MeanParents"]).reshape(-1,1)
y_train_filles = np.array(data_train[data_train["Gender"] == "F"]["Height"]).reshape(-1,1)
reg_train_filles = LinearRegression(fit_intercept=True).fit(x_train_filles, y_train_filles)
y_pred_filles = reg_train_filles.predict(x_test_filles)

# On combine les rÃ©sultats des tests filles / garÃ§ons
y_test = np.concatenate([y_test_garcons, y_test_filles])
y_pred_3 = np.concatenate([y_pred_garcons_2, y_pred_filles])

# On calcule le risque prÃ©dictif total
N = len(y_test)
rPred_combined = 0
for i in range(N):
    rPred_combined+=((y_test[i] - y_pred_3[i])**2)
rPred_combined = rPred_combined/N

print("Le risque de prÃ©diction sur l'Ã©chantillon complet " +
      "avec modÃ¨le diffÃ©rent selon fille/garÃ§on est %f" % rPred_combined)
```

On constate que le risque prÃ©dictif total prenant compte du modÃ¨le fille/garÃ§on selon le genre est bien infÃ©rieur au risque de prÃ©diction basÃ© sur l'Ã©chantillon complet qui ne fait pas la distinction fille/garÃ§on.

```python
# 11.
y_pred_4=[]

N = len(data_test)
rPred_combined_2 = 0
# On parcourt chaque ligne du data set et on calcule les coefficients Ã  chaque fois en utilisant
# le modÃ¨le correspondant au genre fille/garÃ§on correspondant
for index, row in data_test.iterrows():
    isFemale = row['Gender']=='F'
    if(isFemale):
        y_temp = reg_train_filles.predict(np.array(row['MeanParents']).reshape(-1,1))
    else:
        y_temp = reg_train_garcons.predict(np.array(row['MeanParents']).reshape(-1,1))
    y_pred_4.append(y_temp[0][0])
    rPred_combined_2+=((row['Height'] - y_temp[0])**2)
rPred_combined_2 = rPred_combined_2/N
print("Le risque de prÃ©diction sur l'Ã©chantillon complet " +
      "avec modÃ¨le diffÃ©rent selon fille/garÃ§on est %f" % rPred_combined_2[0])
```

On retrouve bien le rÃ©sultat prÃ©cÃ©dent.

```python
# Calcul du coefficient de dÃ©termination
from sklearn.metrics import r2_score
# ModÃ¨le combinÃ© - calcul "Ã  la main"
y_test = data_test['Height']

M=np.ones(len(y_test))*y_test.mean()
v1=(y_pred_4-M)
nominateur=np.dot(v1,v1)
v2=(y_test-M)
denominateur=np.dot(v2,v2)

R2_1=nominateur/denominateur
print("Coefficient de dÃ©termination du modÃ¨le combinÃ©: R2=%f" % R2_1)

#ModÃ¨le initial
R2_2=r2_score(y, y_pred_1)
print("Coefficient de dÃ©termination du modÃ¨le inital: R2=%f" % R2_2)
```

$$R2=\Sigma\dfrac{(\widehat{y_i}-\overline{y})^2}{(y_i-\overline{y})^2}$$

On remarque que le R2 est bien plus Ã©levÃ© pour le modÃ¨le faisant la distinction fille/garÃ§on,.
