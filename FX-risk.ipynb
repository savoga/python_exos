```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
n_days = 100
```

```python
np.random.seed(3)

# Peugeot (EUR)
returns = np.random.normal(0.01,0.01,n_days)
prices_peugeot = 100*np.cumprod(1+returns)

# Roche (CHF)
returns = np.random.normal(-0.01,0.01,n_days)
prices_roche = 100*np.cumprod(1+returns)

# EURCHF
returns_1 = np.random.normal(0.001,0.01,n_days)
rates_curr_1 = .99*np.cumprod(1+returns_1) # EUR is appreciating against CHF
returns_2 = np.random.normal(-0.001,0.01,n_days)
rates_curr_2 = .99*np.cumprod(1+returns_2) # EUR is depreciating against CHF

# Index 50% Peugeot 50% Roche (EUR)
prices_peugeot_eur = prices_peugeot
prices_roche_eur_1 = prices_roche/rates_curr_1
prices_roche_eur_2 = prices_roche/rates_curr_2

prices_index_1 = 0.5*prices_peugeot_eur+0.5*prices_roche_eur_1
prices_index_1 = prices_index_1/prices_index_1[0]*100

prices_index_2 = 0.5*prices_peugeot_eur+0.5*prices_roche_eur_2
prices_index_2 = prices_index_2/prices_index_2[0]*100

# Display

plt.figure(figsize=(10,3))
plt.plot(prices_peugeot)
plt.title('Peugeot (EUR)')
plt.show()

plt.figure(figsize=(10,3))
plt.plot(prices_roche)
plt.title('Roche (CHF)')
plt.show()

plt.figure(figsize=(10,3))
plt.plot(rates_curr_1, label='EUR is appreciating')
plt.plot(rates_curr_2, label='EUR is depreciating')
plt.title('EURCHF')
plt.legend()
plt.show()

plt.figure(figsize=(10,3))
plt.plot(prices_index_1, label='EUR is appreciating')
plt.plot(prices_index_2, label='EUR is depreciating')
plt.legend()
plt.title('Index (EUR)')
plt.show()
```
